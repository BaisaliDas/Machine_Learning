### problem statement 7:
1. extract latitudes & longitudes of banglore place
2. perform geographical analysis (spatial analysis) where exactly most of my rest. are situated

```python
df.shape
```

```python
df['location'].nunique()
```

```python
locations=pd.DataFrame({'Name':df['location'].unique()})
locations
```

```python
lat=[]
lon=[]
for location in locations['Name']:
    location=geolocator.geocode(location)
    if location is None:
        lat.append(np.nan)
        lon.append(np.nan)
    else:
        lat.append(location.latitude)
        lon.append(location.longitude)
```

```python
locations['latitude']=lat
locations['longitude']=lon
```

```python
locations.head()
```

```python
#2
```

```python
Rest_locations=df['location'].value_counts().reset_index()
```

```python
Rest_locations.columns=['Name','count']
Rest_locations
```

```python
Resturant_locations=Rest_locations.merge(locations,on='Name').dropna()
Resturant_locations
```

```python
data=df[df['rest_type']=='Quick Bites']
data.head()
```

```python
data['dish_liked'][3].split()
```

```python
data['dish_liked'].isnull().sum()
```

```python
menu=''
for word in data['dish_liked'].dropna():
    #here i just droped it bcz there is 13843 missing values in this data
    words=word.split() 
    for i in range(len(words)):
        words[i] =words[i].lower()
    menu=menu+' '.join(words)+' '
```

```python
menu
```

- so those need to be pass in wordcloud

```python
stopwords=set(STOPWORDS)
```

```python
wc=WordCloud(stopwords=stopwords,width=1500,height=1500).generate(menu)
```

```python
plt.imshow(wc)
plt.axis('off')
```
